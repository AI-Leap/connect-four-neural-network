```python
#read dataset
df = pd.read_csv('./data/c4_game_database.csv', sep=',',header=0)
df.head(10)
```

```python
cp_callback = tf.keras.callbacks.ModelCheckpoint(filepath = './checkpoints', verbose = 1, save_freq = 'epoch', period = 50)
```

```python
model = model.getModel()

nEpochs = 300
batchSize = 50

history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=nEpochs, batch_size=batchSize, callbacks=[cp_callback])
model.save('./models3')
```

```python
def drawBoard(board):
    nboard = [ [0.0]* 7 for i in range(6)]
    for i in range(6):
        for j in range(7):
            if board[i][j] == 1.0:
                nboard[i][j] = 'X'
            if board[i][j] == -1.0:
                nboard[i][j] = 'O'
            if board[i][j] == 0.0:
                nboard[i][j] = ' '
    return nboard
```
